```python
import pandas as pd
import os
from datetime import datetime
from config import LIQUIDITY_METRICS_CSV, FILTERED_UNIVERSE_CSV


print("Libraries loaded successfully")
```

```python
# Load Data
liquidity_df = pd.read_csv(LIQUIDITY_METRICS_CSV)
print(f"✓ Loaded {len(liquidity_df)} stocks")
```

```python
# Filter and save
filtered_df = liquidity_df[
    (liquidity_df['Avg_Turnover_Cr'] >= 50) &
    (liquidity_df['Median_Turnover_Cr'] >= 30) &
    (liquidity_df['Latest_Price'] >= 300) &
    (liquidity_df['Latest_Price'] <= 4000) &
    (liquidity_df['Avg_Spread_%'] < 2) &
    (liquidity_df['Trading_Days'] >= 100)
].sort_values('Avg_Turnover_Cr', ascending=False).head(15)
```

```python
# Select top 15
print("\n" + "="*80)
print("STEP 4: SELECTING TOP 15 STOCKS")
print("="*80)
top_universe = filtered_df.head(15).reset_index(drop=True)
print(f"✓ Selected top 15 stocks")
print("\nFinal Universe:")
print(top_universe[['Stock', 'Avg_Turnover_Cr', 'Latest_Price', 'Avg_Spread_%']].to_string())
```

```python
top_universe.to_csv(FILTERED_UNIVERSE_CSV, index=False)
```
